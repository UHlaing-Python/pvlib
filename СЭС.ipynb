```python
import cmath

def quadratic(a,b,c):

    d = b**2 - 4*a*c

    if d > 0:
        x1 = (-b + cmath.sqrt(d))/(2*a)
        x2 = (-b - cmath.sqrt(d))/(2*a)
        print("The results of x are:", x1, "and", x2)

    elif d == 0:
        x1 = -b/2*a
        print("The result of x is:", x1)

    else:
        print("No real values are there")

a = float(input("Enter value of a!"))
b = float(input("Enter value of b!"))
c = float(input("Enter value of c!"))

if a == 0:
    print("Incorrect Equation!")
else:
    quadratic(a,b,c)
```

```python
import pvlib
```

```python
from pvlib.location import Location
```

```python
A = Location(55.4, 37.4, tz = "Europe/Moscow", altitude = 156, name = "Moscow")
print(A)
```

```python
import pandas as pd
import datetime
```

```python
times = pd.date_range(start = datetime.datetime(2021,3,23), end = datetime.datetime(2021,3,23,23), freq = "1H")
```

```python
times
```

```python
times_loc = times.tz_localize(A.pytz)
times_loc
```

```python
SPA = pvlib.solarposition.spa_python(times_loc, A.latitude, A.longitude)
```

```python
SPA
```

```python
import matplotlib.pyplot as plt
```

```python
SPA["zenith"].plot(label = "zenith")
SPA["elevation"].plot(label = "elevation")
SPA["azimuth"].plot(label = "azimuth")
plt.grid()
plt.legend()
plt.title("ПОложение Солца")             
```

```python
DNI_extra = pvlib.irradiance.get_extra_radiation(times_loc, method = "spencer")
```

```python
DNI_extra 
```

```python
apparent_elevation = SPA["apparent_elevation"]
VerySimpleClearSky = pvlib.clearsky.simplified_solis(apparent_elevation, aod700=0.1, precipitable_water=1.0, pressure=101325.0, dni_extra=1364.0)
VerySimpleClearSky
```

```python
VerySimpleClearSky.plot()
plt.legend()
plt.ylabel("Иррадиация")
```

```python
apparent_zenith = SPA["apparent_zenith"]
zenith = SPA["zenith"]
airmass_relative = pvlib.atmosphere.get_relative_airmass(zenith, model='simple')
airmass_absolute = pvlib.atmosphere.get_absolute_airmass(airmass_relative, pressure=101325.0)
linke_turbidity = pvlib.clearsky.lookup_linke_turbidity(times_loc, A.latitude, A.longitude, filepath=None, interp_turbidity=True)
SimpleClearSky = pvlib.clearsky.ineichen(apparent_zenith, airmass_absolute, linke_turbidity, altitude=0, dni_extra=DNI_extra, perez_enhancement=False)
SimpleClearSky
```

```python
SimpleClearSky.plot()
plt.legend()
plt.ylabel("Иррадиация")
```

```python
SimpleClearSky["ghi"].plot(label = "Простая Модель")
VerySimpleClearSky['ghi'].plot(label = "Простейшая Модель")
plt.legend()
```
